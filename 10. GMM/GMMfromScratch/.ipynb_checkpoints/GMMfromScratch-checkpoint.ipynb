### Gaussian Mixture Model (GMM) in Simple English:

A Gaussian Mixture Model (GMM) is a machine learning algorithm used to group data into different clusters. It assumes that the data is made up of a mixture of several Gaussian Distributions (also called Normal Distributions).


--

### How GMM Works Step-by-Step:

1. What is Gaussian Distribution?
   - It is a bell-shaped curve that represents how data points are distributed around the mean.
   - Example: Heights of people in a classroom or marks in an exam often follow this distribution.

2. What is Mixture Model?
   - When data contains multiple Gaussian Distributions combined together, it is called a Mixture Model.
   - Example: Suppose you have flowers in different shapes and sizes. Each type of flower might follow a different Gaussian distribution.

3. How GMM Works:
   - First, GMM assumes how many clusters are present in the data (e.g., 3 clusters for Iris dataset).
   - It assigns random means (centers), covariances (spread of data), and weights (importance) to each cluster.

4. Expectation-Maximization (EM Algorithm):
   - E-Step (Expectation): Calculates the probability that each point belongs to a particular cluster.
   - M-Step (Maximization): Updates the cluster means, covariances, and weights based on the calculated probabilities.
   - This process repeats until the model becomes stable.

5. Advantages:
   - Can find hidden patterns in data.
   - Soft clustering (a point can belong to more than one cluster with different probabilities).

6. Disadvantages:
   - Can be slow for large datasets.
   - Sensitive to wer might belong to multiple groups.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
# Load Iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]  # Using first two features for easy visualization
y = iris.target
```

```python
# Standardize data
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
class GMM:
    def __init__(self, n_components, n_iter=100, tol=1e-4):
        self.n_components = n_components
        self.n_iter = n_iter
        self.tol = tol
        self.weights = None
        self.means = None
        self.covariances = None

    def initialize(self, X):
        n_samples, n_features = X.shape
        self.weights = np.ones(self.n_components) / self.n_components
        self.means = X[np.random.choice(n_samples, self.n_components, replace=False)]
        self.covariances = np.array([np.cov(X.T) for _ in range(self.n_components)])

    def gaussian(self, X, mean, cov):
        d = X.shape[1]
        coeff = 1 / (np.sqrt((2 * np.pi) ** d * np.linalg.det(cov)))
        exp_part = np.exp(-0.5 * np.sum(np.dot(X - mean, np.linalg.inv(cov)) * (X - mean), axis=1))
        return coeff * exp_part

    def e_step(self, X):
        responsibilities = np.zeros((X.shape[0], self.n_components))
        for k in range(self.n_components):
            responsibilities[:, k] = self.weights[k] * self.gaussian(X, self.means[k], self.covariances[k])
        responsibilities /= responsibilities.sum(axis=1, keepdims=True)
        return responsibilities

    def m_step(self, X, responsibilities):
        Nk = responsibilities.sum(axis=0)
        self.weights = Nk / X.shape[0]
        self.means = np.dot(responsibilities.T, X) / Nk[:, np.newaxis]
        for k in range(self.n_components):
            diff = X - self.means[k]
            self.covariances[k] = np.dot(responsibilities[:, k] * diff.T, diff) / Nk[k]
            self.covariances[k] += 1e-6 * np.eye(X.shape[1])  # To avoid singular matrix

    def fit(self, X):
        self.initialize(X)
        log_likelihood = 0
        for _ in range(self.n_iter):
            responsibilities = self.e_step(X)
            self.m_step(X, responsibilities)
            new_log_likelihood = np.sum(np.log(responsibilities.sum(axis=1)))
            if abs(new_log_likelihood - log_likelihood) < self.tol:
                break
            log_likelihood = new_log_likelihood

    def predict(self, X):
        responsibilities = self.e_step(X)
        return np.argmax(responsibilities, axis=1)
```

```python
# Train GMM model
gmm = GMM(n_components=3, n_iter=100)
gmm.fit(X_train)
y_pred = gmm.predict(X_test)
```

```python
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Plot actual vs predicted
plt.figure(figsize=(12, 12))
plt.subplot(1, 3, 1)
plt.title("Actual Classes")
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='viridis')

plt.subplot(1, 3, 2)
plt.title("Predicted Classes")
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='viridis')

# Decision Boundary
plt.subplot(1, 3, 3)
plt.title("Decision Boundary")
x_min, x_max = X[:, 0].min() - 0.2, X[:, 0].max() + 0.2
y_min, y_max = X[:, 1].min() - 0.2, X[:, 1].max() + 0.2
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = gmm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.6, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')

plt.show()
```
