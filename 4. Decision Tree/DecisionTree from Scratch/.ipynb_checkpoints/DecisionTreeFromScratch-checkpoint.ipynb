```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
# Load Iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]  # Use first two features for visualization
y = iris.target
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Node class for the tree
class Node:
    def __init__(self, feature=None, threshold=None, left=None, right=None, value=None):
        self.feature = feature
        self.threshold = threshold
        self.left = left
        self.right = right
        self.value = value
```

### Ginni Impurity
- Calculates the probability of each label.
- Subtracts the sum of squared probabilities from 1.
- Lower result means purer labels.

```python
# Gini Impurity
def gini(y):
    m = len(y)
    if m == 0:
        return 0
    _, counts = np.unique(y, return_counts=True)
    p = counts / m
    return 1 - np.sum(p ** 2)
```

### Split dataset based on threshold
- Splits the dataset into two parts based on the threshold.
- Left group: Feature values â‰¤ threshold.
- Right group: Feature values > threshold.

```python
# Split dataset based on threshold
def split(X, y, feature, threshold):
    left_idx = X[:, feature] <= threshold
    right_idx = X[:, feature] > threshold
    return X[left_idx], X[right_idx], y[left_idx], y[right_idx]
```

### Best Split
- Tries every feature and possible threshold.
- Splits data and calculates Gini impurity.
- Picks the feature and threshold with the lowest impurity.

```python
# Best Split
def best_split(X, y):
    m, n = X.shape
    best_gini = 1
    best_feature, best_threshold = None, None
    for feature in range(n):
        thresholds = np.unique(X[:, feature])
        for threshold in thresholds:
            X_left, X_right, y_left, y_right = split(X, y, feature, threshold)
            gini_split = (len(y_left) * gini(y_left) + len(y_right) * gini(y_right)) / len(y)
            if gini_split < best_gini:
                best_gini = gini_split
                best_feature = feature
                best_threshold = threshold
    return best_feature, best_threshold
```

### Build Tree
- Stops splitting if labels are pure or maximum depth is reached.
- Splits data using the best feature and threshold.
- Recursively builds left and right branches.

```python
# Build Tree
def build_tree(X, y, depth=0, max_depth=3):
    if len(np.unique(y)) == 1 or depth == max_depth:
        leaf_value = np.argmax(np.bincount(y))
        return Node(value=leaf_value)
    feature, threshold = best_split(X, y)
    X_left, X_right, y_left, y_right = split(X, y, feature, threshold)
    left = build_tree(X_left, y_left, depth + 1, max_depth)
    right = build_tree(X_right, y_right, depth + 1, max_depth)
    return Node(feature, threshold, left, right)
```

```python
# Predict
def predict_one(node, x):
    if node.value is not None:
        return node.value
    if x[node.feature] <= node.threshold:
        return predict_one(node.left, x)
    else:
        return predict_one(node.right, x)
```

```python
def predict(tree, X):
    return np.array([predict_one(tree, x) for x in X])
```

```python
# Train model
tree = build_tree(X_train, y_train)
y_pred = predict(tree, X_test)
```

```python
# Accuracy
accuracy = np.sum(y_pred == y_test) / len(y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
# Plot Decision Boundary
def plot_decision_boundary(X, y, model):
    cmap = ListedColormap(['#FFAAAA', '#AAAAFF', '#AAFFAA'])
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))
    Z = predict(model, np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=cmap)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', cmap=cmap)
    plt.title('Decision Tree Decision Boundary')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

plot_decision_boundary(X, y, tree) 
```

```python
# Actual vs Predicted
print("Actual vs Predicted:")
print(pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}))
```
