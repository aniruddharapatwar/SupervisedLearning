```python
import pandas as pd
import numpy as np
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:,2:4]
y = iris.target
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(multi_class='ovr',solver='lbfgs',max_iter=200)
log_reg.fit(X,y)
```

```python
x_min, x_max = X[:,0].min() - 0.5, X[:,0].max() + 0.5
y_min, y_max = X[:,0].min() - 0.5, X[:,0].max() + 0.5
```

```python
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
```

```python
Z = log_reg.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
# 6. Plot Decision Boundaries + Scatter Plot
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.coolwarm)
plt.title("Logistic Regression Decision Boundaries on Iris Data")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()
```
